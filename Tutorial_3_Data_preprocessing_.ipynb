```python
import pandas as pd
```

```python
#The UCI (University of California, Irvine) Machine Learning Repository is a popular online collection of datasets.

#Load the automobile data from UCI data repository
url = 'https://archive.ics.uci.edu/ml/machine-learning-databases/autos/imports-85.data'

column_names = ['symboling', 'normalized-losses', 'make', 'fuel-type', 'aspiration', 'num-of-doors',
                'body-style', 'drive-wheels', 'engine-location', 'wheel-base', 'length', 'width', 'height',
                'curb-weight', 'engine-type', 'num-of-cylinders', 'engine-size', 'fuel-system',
                'bore', 'stroke', 'compression-ratio', 'horsepower', 'peak-rpm', 'city-mpg',
                'highway-mpg', 'price']

#for null values see as Non add "  na_values='?'    "
df = pd.read_csv(url, names=column_names , na_values='?')
df.head()
```

```python
print(df.shape)
```

```python
df.info()
```

```python
df.describe()
```

```python
# Study the data type of each feature
data_types = df.dtypes
print(data_types)
```

```python
# Print a report on the number of null values in each column
null_counts = df.isnull().sum()
print(null_counts)
```

```python
#Find the null values in the column ’number of doors’.

null_doors = df[df['num-of-doors'].isnull()]
null_doors
```

```python
# instances mean essentially counting the number of occurrences of a specific value in a column or dataset.

#In the same column, count the number of instances in each value:
door_counts = df['num-of-doors'].value_counts()
door_counts
```

```python
# Specify the columns you want to print
column_to_print = ['num-of-doors']
rows_to_print = [27, 63]

# Print the specified columns and raw
print(df.loc[rows_to_print, column_to_print])
```

```python
# Replace the null values in that column with the number of doors with maximum instances:
max_door_count = door_counts.idxmax()
df['num-of-doors'].fillna(max_door_count, inplace=True)

# Print the specified columns and raw
print(df.loc[rows_to_print, column_to_print])
```

```python
df.head(5)
```

```python
# Select columns with categorical data types
categorical_columns = df.select_dtypes(include=['object', 'category']).columns

# Print the categorical columns
categorical_columns
```

```python
#Change all the categorical values into numeric values:

from sklearn.preprocessing import LabelEncoder

# Apply Label Encoding to each categorical column
label_encoder = LabelEncoder()
for column in categorical_columns:
    df[column] = label_encoder.fit_transform(df[column].astype(str))
```

```python
df.head(5)
```

```python
# To identify the columns with null values
null_columns = df.columns[df.isnull().any()]
print(null_columns)
```

```python
print(df.loc[:, null_columns])
```

```python
# Replace "NaN" values in numeric columns with the mean of the column

import numpy as np

numeric_columns = df.select_dtypes(include=np.number).columns
df[numeric_columns] = df[numeric_columns].fillna(df[numeric_columns].mean())

 # first identify the numeric columns using select_dtypes() with np.number as the argument.
 # We then replace the "NaN" values in those columns with the mean of the column using the fillna() method.
```
