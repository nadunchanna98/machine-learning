# Boolean masking technique

```python
data = {"Name":["nadun","channa","perera","Ruvi","Ruvindya","sachi"],"Age":[20,52,19,51,95,10],"Gender":["male","female","male","male","female","male"]}
df = pd.DataFrame(data)
df
```

```python
df["Age"]>50
```

```python
df[df["Age"]>50]
```

```python
df[(df["Age"]>50) & (df["Gender"] == "female")]
```

# Summaries for other variables with grouping

```python
data = {"ID":[1258,5254,1779,5361,9795,2120],"Name":["nadun","channa","perera","Ruvi","Ruvindya","sachi"],"Age":[20,52,19,51,95,20],"Gender":["m","f","m","m","f","m"],"Marks":[60,2,91,100,60,14]}
df = pd.DataFrame(data)
df
```

```python
#first remove the id column.because it is not need
df.drop("ID",axis=1,inplace=True)
```

```python
df
```

```python
df.groupby("Gender").mean()
```

```python
#get like panda series
df.groupby("Gender")["Marks"].mean()
```

# Concatenating data frames

```python
data = {"A":[1258,5254,1779,5361,9795,2120],"B":[20,52,19,51,95,20]}
df1 = pd.DataFrame(data)
df1
```

```python
data = {"A":[58,54,79,61,95,20],"B":[2120,5242,1559,5451,9755,2074]}
df2 = pd.DataFrame(data)
df2
```

# Concatenating along the rows

```python
pd.concat([df1,df2])
```

```python
############## ignore index
pd.concat([df1,df2],ignore_index=True)
```

# Concatenating along the Colomns

```python
pd.concat([df1,df2],axis=1)
```

```python
#another way "loc".its like "iloc" . but iloc cannot use
df1.loc[6]=[1000,3333]
df1
```
