```python
import csv

with open('/content/drive/MyDrive/RESEACH/DATASET1/identifiers.csv', 'r') as file1:
    reader1 = csv.DictReader(file1)

    # Create a dictionary to store the data from the first CSV file
    dict1 = {}
    for row in reader1:
        key = row['sample']
        dict1[key] = row
```

```python
import csv

with open('/content/drive/MyDrive/RESEACH/DATASET1/features.csv', 'r') as file2:
    reader2 = csv.DictReader(file2)

  # Create a list to store the joined data
    joined_data = []

    # Create a dictionary to store the data from the first CSV file
    for row in reader2:
        key = row['sampleID']

           # If the key exists in the first CSV file, join the rows
        if key in dict1:

            # Create a new dictionary with columns from both CSV files
            joined_row = {}
            
            for col in dict1[key]:
                joined_row[col] = dict1[key][col]
            for col in row:
                if col not in dict1[key]:
                    joined_row[col] = row[col]

            # Add new columns as needed
            joined_row['new_column'] = 'new_value'
            joined_data.append(joined_row)

# Write the joined data to a new CSV file
with open('/content/drive/MyDrive/RESEACH/DATASET1/joined.csv', 'w', newline='') as joined_file:
    fieldnames = list(joined_data[0].keys())
    writer = csv.DictWriter(joined_file, fieldnames=fieldnames)
    writer.writeheader()
    for row in joined_data:
        writer.writerow(row)            
```
